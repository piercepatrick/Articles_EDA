```python
g20 = ['Argentina', 'Austrailia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'India', 'Indonesia', 'Italy', 'Japan', 'Republic of Korea', 'Mexico', 'Russia', 'Saudi Arabia', 'South Africa', 'Turkey', 'United Kingdom', 'United States', 'European Union']
# Create binary columns for each country
# Loop through each row and if the word is in the article, assign the value 1
def countryInArticle(row):
    if country in row:
        return 1
    else:
        return 0
for country in g20:
    df[country] = df['text'].apply(countryInArticle)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
# Creating a column chart of the count of the records
sns.countplot(x = "genre",  data = df, color = "royalblue")
# You can see color chocies at https://python-graph-gallery.com/100-calling-a-color-with-seaborn/

# Create a label for the x axis
plt.xlabel('Number of Articles per Genre', fontweight='bold', fontsize='14', horizontalalignment='center')

# Create a label for the y axis in the chart
plt.ylabel('Count', fontweight='bold', fontsize='14', horizontalalignment='center')

plt.show()
```

```python
df_sum = df.sum(numeric_only = True)
df_gby = df.groupby(['genre']).sum(numeric_only = True)
```

```python
# Set style and size of plot
sns.set(style='whitegrid')
plt.figure(figsize=(16,6))

# Define the chart as a count of number of observations, with pre-determined order
chart = df_sum.plot.bar(color = "navy")

# Set the size and alignment of the x axis labels and rotate them
chart.set_xticklabels(chart.get_xticklabels(), 
                      fontsize='12', rotation=45, 
                      horizontalalignment='right')

plt.suptitle('    Frequency of Countries mentioned in Articles', fontweight='bold', 
          fontsize='20', horizontalalignment='center', y=.98)


# Create labels for the x and y axes
plt.xlabel('Country', fontweight='bold', 
           fontsize='14', horizontalalignment='center')

plt.ylabel('Number of Articles Mentioned In', fontweight='bold', 
           fontsize='14', horizontalalignment='center')
plt.savefig('ArticlesGraph1.png')
plt.show()
```

```python
# Set style and size of plot
sns.set(style='whitegrid')


# Define the chart as a count of number of observations, with pre-determined order
chart = df_gby.transpose().plot.bar(stacked=True)

# Set the size and alignment of the x axis labels and rotate them
chart.set_xticklabels(chart.get_xticklabels(), 
                      fontsize='12', rotation=45, 
                      horizontalalignment='right')

plt.suptitle('    Distribution of Countries in Articles', fontweight='bold', 
          fontsize='20', horizontalalignment='center', y=1.05)
plt.title('Stacked by Genre', fontweight="bold", 
          fontsize='15', horizontalalignment='center', y=1.02)

# Create labels for the x and y axes
plt.xlabel('Country', fontweight='bold', 
           fontsize='14', horizontalalignment='center')

plt.ylabel('Count', fontweight='bold', 
           fontsize='14', horizontalalignment='center')
plt.savefig('ArticlesGraph2.png')
plt.show()
```

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Generate wordcloud for 'description' column
words = " ".join(df['text'])

def punctuation_stop(text):
    """remove punctuation and stop words"""
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = punctuation_stop(words)

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 2000, width =800, height = 1500)
wc.generate(text)

plt.figure(figsize=[15,10])
plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.savefig('ArticlesWordCloud.png')
plt.show()
```
